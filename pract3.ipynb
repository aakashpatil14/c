```python
import matplotlib.pyplot as plt 
```

```python
import numpy as np
```

```python
def fun(x):
    return (x+3)**2
```

```python
def gradient1(x):
    return 2*(x+3)
```

```python
def gradient_descent(fun,start,n_iter,learning_rate):
    function=fun
    gradient=gradient1
    points=[start]
    iter=0
    while iter<=n_iter:
        prev=start
        start=start-learning_rate*(gradient(prev))
        points.append(start)
        iter=iter+1
    print("local minima occurs at ",start)  

    x_=np.linspace(-7,5,100)
    y=function(x_)
    fig=plt.figure(figsize=(10,10))
    plt.plot(x_,y,'g')
    plt.plot(points,function(np.array(points)),'-o')
    plt.show()
    

    
```

```python
gradient_descent(fun,2.0,100,0.2)
```
